```python
import numpy as np
def T_NH3(P,Z): 
    return 1e4 / (68.02 - 6.19 * Z - 6.31 * np.log10(P)) # Lodders & Fegley (2002)

def T_H2O(P,Z):
    return 1e4 / (38.84 - 3.93 * Z - 3.83 * np.log10(P) - 0.20 * Z * np.log10(P)) # Lodders & Fegley (2002)

```

```python
from exojax.atm.psat import psat_ammonia_AM01, psat_water_AM01
```

```python
import numpy as np
from exojax.utils.zsol import nsol

nsun = nsol()
print(nsun['Fe'], np.log10(nsun["Fe"]), nsun["H"], nsun["He"], np.log10(nsun["N"]))
```

```python
import matplotlib.pyplot as plt
#pressure = np.logspace(-5,2,100)
tarr = np.linspace(150,500,100)
p = psat_water_AM01(tarr)#/nsun["O"]

parr = np.logspace(-8,5,100)
fig = plt.figure()
ax = fig.add_subplot(111)
plt.plot(tarr,p)
plt.plot(T_H2O(parr,0.0),parr*nsun["O"], label="Z=0")
#plt.plot(T_H2O(parr,1.0),parr, label="Z=1")
plt.legend()
plt.axvline(273.+100.0)
plt.axhline(1.0)

plt.yscale("log")
plt.xscale("log")
#ax.invert_yaxis()
```

```python
import matplotlib.pyplot as plt
#pressure = np.logspace(-5,2,100)
tarr = np.logspace(1.9,2.5,100)
p = psat_ammonia_AM01(tarr)

parr = np.logspace(-8,5,100)
fig = plt.figure()
ax = fig.add_subplot(111)
plt.plot(tarr,p)
plt.plot(T_NH3(parr,0.0),parr*nsun["N"])
plt.yscale("log")
plt.xscale("log")
ax.invert_yaxis()
```
