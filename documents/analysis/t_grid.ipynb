## The reason why we use E grid instead of T

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
Tarr = np.linspace(100,5000, 100)
Earr = np.logspace(2,4.,1000)
```

```python
from exojax.utils.constants import hcperk, Tref
def East(T):
    return -1. / (hcperk * (1. / T - 1. / Tref))

def f(E,T):
    return np.exp(E/East(T))
```

```python
plt.plot(Earr,f(Earr,1100)/f(Earr,1000),label="$f(E,1100K)/f(E,1000K)$")
plt.plot(Earr,f(Earr,900)/f(Earr,1000),label="$f(E,900K)/f(E,1000K)$",ls="dashed")
plt.plot(Earr,f(Earr,600)/f(Earr,500),label="$f(E,600K)/f(E,500K)$")
plt.plot(Earr,f(Earr,400)/f(Earr,500),label="$f(E,400K)/f(E,500K)$",ls="dashed")
plt.ylim(0.0,4.0)
plt.legend()
plt.axhline(1.0,color="gray")
plt.xlabel("E ($\mathrm{cm}^{-1}$)")
plt.xscale("log")
```

```python
plt.plot(Tarr,f(1100,Tarr)/f(1000,Tarr),label="$f (E+100\mathrm{cm^{-1}},T) / f (E,T)$")
plt.plot(Tarr,f(900,Tarr)/f(1000,Tarr),label="$f (E-100\mathrm{cm^{-1}},T) / f (E,T)$",ls="dashed")
#plt.plot(Tarr,s(200,Tarr)/s(300,Tarr),label="")
#plt.plot(Tarr,s(400,Tarr)/s(300,Tarr),label="",ls="dashed")
#does not depend on E but Delta E
plt.legend()
plt.axhline(1.0,color="gray")
plt.axvline(296,alpha=0.3,color="red")
plt.text(296+10,2.,"$T_\mathrm{ref}=296K$",color="red",rotation="vertical",alpha=0.6)
plt.xlabel("T (K)")
plt.xscale("log")
```
