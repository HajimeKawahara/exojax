```python
from exojax.spec.hitrancia import read_cia
import jax.numpy as jnp
from jax import jit

import numpy as np
import matplotlib.pyplot as plt
```

```python
nucia,tcia,ac=read_cia("../../data/CIA/H2-H2_2011.cia",1.e7/1645.,1.e7/1630.0)
```

```python
len(nucia),np.shape(ac),len(tcia)
```

```python
logac=jnp.array(np.log10(ac))
```

```python
plt.plot(tcia,logac[:,0])
```

```python
@jit
def fcia(x,i):
    return jnp.interp(x,tcia,logac[:,i])
```

```python
fcia(2000.,33)
```

```python
plt.plot(tcia,logac[:,0],".")
plt.plot(tcia,fcia(tcia,0))
plt.plot(tcia,logac[:,-1],".")
plt.plot(tcia,fcia(tcia,-1))
```
